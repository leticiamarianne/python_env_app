#### Vazão de escoamento de chuvas

Q = C * I * A

```python
import numpy as np
import matplotlib.pyplot as plt

# considerando áreas urbanas: C de 0,5 a 0,8
val_min_c = 0.5
val_max_c = 0.9
array_coef_aleat = (val_max_c - val_min_c) * np.random.rand(100) + val_min_c
print(array_coef_aleat)

# considerando intensidade variando de 5 a 100 mm/h
val_min_i = 5
val_max_i = 100
array_intens_aleat = (val_max_i - val_min_i) * np.random.rand(100) + val_min_i
print(array_intens_aleat)

# considerando área variando de 1 a 50 km²
val_min_a = 0.5
val_max_a = 0.9
array_area_aleat = (val_max_a - val_min_a) * np.random.rand(100) + val_min_a
print(array_area_aleat)
```

```python
Q = array_coef_aleat * array_intens_aleat * array_area_aleat
print(Q)

#convertendo unidades
conversao = 1000/3600
Q_m3s = Q * conversao
print(Q_m3s)
```

```python
# Criando subplots

fig,ax = plt.subplots()
ax.plot(Q_m3s)

# 1 linha e 3 colunas/gráficos
fig,ax = plt.subplots(1, 3, figsize=(15,5))

# Vazão vs coeficiente
ax[0].scatter(array_coef_aleat, Q_m3s, color='blue')
ax[0].set_xlabel('Coeficiente de escoamento')
ax[0].set_ylabel('Vazão (m³/s')
ax[0].set_title('Vazão vs Coeficiente')

# Vazão vs intensidade
ax[1].scatter(array_intens_aleat, Q_m3s, color='red')
ax[1].set_xlabel('Intensidade (mm/h)')
ax[1].set_ylabel('Vazão (m³/s')
ax[1].set_title('Vazão vs Intensidade')

#Vazão vs área
ax[2].scatter(array_area_aleat, Q_m3s, color='orange')
ax[2].set_xlabel('Área (km²)')
ax[2].set_ylabel('Vazão (m³/s')
ax[2].set_title('Vazão vs Área')
```

#### Equação de risco
Risco = Ameaça * Vulnerabilidade

```python
# forma equivalente

n = 100

ameaca = np.random.uniform(0, 1, n)
vulnerabilidade = np.random.uniform(0, 1, n)

risco = ameaca * vulnerabilidade
print(risco)

fig,ax = plt.subplots()
ax.plot(risco)

fig,ax = plt.subplots(1, 2, figsize=(8,3))

ax[0].scatter(ameaca, risco, color='steelblue', s=10)
ax[0].set_xlabel('Ameaça', fontsize=12, fontname='Times New Roman')
ax[0].set_ylabel('Risco', fontsize=12, fontname='Times New Roman')
ax[0].set_title('Risco em função da Ameaça', fontsize=14, fontname='Times New Roman')

ax[1].scatter(vulnerabilidade, risco, color='crimson', s=10)
ax[1].set_xlabel('Vulnerabilidade', fontsize=12, fontname='Times New Roman')
ax[1].set_ylabel('Risco', fontsize=12, fontname='Times New Roman')
ax[1].set_title('Risco em função da Vulnerabilidade', fontsize=14, fontname='Times New Roman')

# alterando a fonte dos números dos eixos
for axis in ax:
    for label in axis.get_xticklabels() + axis.get_yticklabels():
        label.set_fontname('Times New Roman')
```
