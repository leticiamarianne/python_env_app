```python
import pandas as pd
import os
```

```python
pasta = r'C:\Users\leti_\OneDrive\Documentos\python_env_app\MTR'
arquivos = [ fname for fname in os.listdir(pasta) if fname.endswith('.csv')]

#df = pd.read_csv(r"C:\Users\leti_\OneDrive\Documentos\python_env_app\MTR\dados_MTR.csv")

# Criando DataFrame
df = []

for arq in arquivos:
    temp = pd.read_csv(os.path.join(pasta, arq), sep=';')
    df.append(temp)

# Concatenando em um DataFrame
df = pd.concat(df, ignore_index=True)

df.head()
```

```python
# 2. Converter a coluna de data
# (troque 'data' pelo nome exato da coluna de datas da sua planilha)
df['Data'] = pd.to_datetime(df['Data'])

# 3. Criar colunas de ano e mês
df['ano'] = df['Data'].dt.year
df['mes'] = df['Data'].dt.month

df.head()
```
