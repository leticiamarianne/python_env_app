### Listas

```python
# primeira lista
lista_a = ['Igor', 'Clarissa', 'Leonardo']
print(lista_a)

lista_b = ['José', 7, 6.5]
print(lista_b)
print(type(lista_b))

#indexação
#primeiro valor da lista b
print(lista_b[0])
print(lista_b[0] + 'tem ' + str(lista_b[1]) + ' meses e ' + str(lista_b[2]) + ' kg')

print(type(lista_b[0]))
print(type(lista_b[1]))
```

```python
# criando uma lista de valores consecutivos
lista_consecutivos = list(range(0,100))
print(lista_consecutivos)

# primeiro valor
lista_consecutivos[0]
print(lista_consecutivos[0])

# último valor
lista_consecutivos[99]
print(lista_consecutivos[99])
print(lista_consecutivos[-1])

#comprimento da lista
comprimento = len(lista_consecutivos)
print(comprimento)

# valor no meio da série
metade = comprimento/2
print(metade)
print(lista_consecutivos[int(metade)])

# dica: se escrever a primeira letra do comando e apertar tab aparece todos os comandos com aquela letra
```

### Loop

```python
# usando range

for ii in range(0,10):
    print(ii)

# usando uma lista e range
for ii in range(0,len(lista_a)):
    print(lista_a[ii])
```

```python
# exemplo

dic_2 = {'Igor': 10,
         'Marcos': 9,
         'Júlia': 5.75,
         'Clarrissa': 5.75,
         'Leonardo': 4,
         'Leticia': 8}

for key,value in dic_2.items():
    if(value>8):
        print(key + ' está aprovado com notão')
    elif value == 5.75:
            print(key + ' passou raspando')
    elif value > 5.75:
            print(key + ' está aprovado')
    else:
            print(key + ' está reprovado')
```

```python
# Loop em uma lista e acumulando valores específicos

lista_c = [100, 200, 10, 2, 1, 6, 22, 27]

for index,valor in enumerate(lista_c):
    print('índice da lista = ' + str(index))

#acumulando valores maiores do que 50
lista_maior_50 = []

for index,valor in enumerate(lista_c):
    if valor>50:
        print('índice da lista = ' + str(index))

lista_maior_50.append(valor)
print(lista_maior_50)
```

```python
# usando lógica e numpy

#importando o numpy
import numpy as np

# convertendo para numpy
arr = np.array(lista_c)
print(arr)
print(type(arr))

#selecionando valores maiores que 50 - retorna um array booleano
print(arr>50)

# imprimir apenas valores maiores que 50
print(arr[arr>50])
```

```python
# média, mínimo, máximo, desvio, percentil mediana

print('Média = ' + str(arr.mean()))
print('Mínimo = ' + str(arr.min()))
print('Máximo = ' + str(arr.max()))
print('Percentil90 = ' + str(np.percentile(arr,90)))
print('Percentil50 = ' + str(np.percentile(arr,50)))
print('Mediana = ' + str(np.median(arr)))
```

```python
# array de números aleatórios

val_min = 80
val_max = 600

array_temperatura_aleatoria = (val_max - val_min) * np.random.rand(100) + val_min
print(array_temperatura_aleatoria)
```

```python
# exemplo de cinética

fx = 200 * np.exp(array_temperatura_aleatoria) + array_temperatura_aleatoria**(-0.2)
print(fx)
```

### Figuras - Matplotlib

```python
# importando o pacote
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.plot(fx)
# plot é mais para séries temporais

fig,ax = plt.subplots(2)
ax[0].plot(fx)
ax[1].scatter(array_temperatura_aleatoria,fx,color='r')
ax[1].set_ylabel('fx')
```
